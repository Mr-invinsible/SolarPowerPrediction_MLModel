```python
import matplotlib.pyplot as plt
```

```python
# plot distribution of power
plt.figure(figsize=(10,6))
sns.histplot()
```

```python
plt.Figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'], bins=30, kde=True)
plt.title('distribution of generated power (kw)')
plt.xlabel('Generated power (kw)')
plt.ylabel('frequency')
plt.show()
```

```python
# Histogram: Distribution of temperature
plt.figure(figsize=(10, 6))
sns.histplot(df['temperature_2_m_above_gnd'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Temperature (2m above ground)')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter Plot: Generated power vs wind speed
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='wind_speed_10_m_above_gnd', y='generated_power_kw', alpha=0.7, color='green')
plt.title('Generated Power vs Wind Speed (10m above ground)')
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Generated Power (kW)')
plt.show()
```

```python
# Box Plot: Total precipitation by cloud cover levels
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='total_cloud_cover_sfc', y='total_precipitation_sfc', palette='Set3')
plt.title('Total Precipitation vs Cloud Cover')
plt.xlabel('Total Cloud Cover (%)')
plt.ylabel('Total Precipitation (mm)')
plt.show()
```

```python
# Line Plot: Wind speed and power generation
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='wind_speed_80_m_above_gnd', y='generated_power_kw', color='orange')
plt.title('Generated Power vs Wind Speed (80m above ground)')
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Generated Power (kW)')
plt.show()
```

```python
# Heatmap: Correlation matrix
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix of Features')
plt.show()
```

```python
# Pairplot: Relationships between key variables
sns.pairplot(df, vars=[
    'temperature_2_m_above_gnd', 
    'wind_speed_10_m_above_gnd', 
    'generated_power_kw', 
    'relative_humidity_2_m_above_gnd'
], diag_kind='kde')
plt.suptitle('Pairplot of Key Variables', y=1.02)
plt.show()
```

```python
# Violin Plot: Wind gust distribution by power generation levels
df['power_category'] = pd.cut(df['generated_power_kw'], bins=3, labels=['Low', 'Medium', 'High'])
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='power_category', y='wind_gust_10_m_above_gnd', palette='muted')
plt.title('Wind Gust Distribution by Power Generation Levels')
plt.xlabel('Power Category')
plt.ylabel('Wind Gust (m/s)')
plt.show()
```

```python
# Bar Plot: Average generated power by cloud cover levels
avg_power_by_cloud_cover = df.groupby('total_cloud_cover_sfc')['generated_power_kw'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_power_by_cloud_cover, x='total_cloud_cover_sfc', y='generated_power_kw', palette='viridis')
plt.title('Average Generated Power by Cloud Cover Levels')
plt.xlabel('Total Cloud Cover (%)')
plt.ylabel('Average Generated Power (kW)')
plt.show()
```
